```python
pd.read_sql_query("SELECT count(start_station_name) as count_station, start_station_id, start_station_name as most_popular_station FROM trip group by(start_station_name) order by count_station DESC limit 20", conn)
```

```python
pd.read_sql_query("SELECT count(*) as idle_time, station_id FROM status where station_id = 2 and bikes_available > 3", conn)
```
